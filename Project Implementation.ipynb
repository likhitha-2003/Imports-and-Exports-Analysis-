### Data Cleaning

```python
trade.isnull() # detects missing values in the given series object
```

```python
trade.isnull().sum() # counts the no of missing values in each row 
```

```python
# To print the only null values records
trade[trade.isnull().any(axis=1)]
# axis=1 for performing action a column
```

```python
# notnull()
trade.notnull() # for non-missing values
```

```python
# To count the not null records
trade.notnull().sum()
```

```python
# To print the not null records.
trade[trade.notnull().all(axis=1)]
```

```python
# Percentage of missing values round by two digit after decimal.
#Reason behind to check whether it can be droped.
round(trade.isnull().mean()*100,2)
```

#### dropna(): TO remove the null values..

```python
# Drop null
trade = trade.dropna()
trade.head()
```

```python
trade.dtypes # before conversion these are the data types
```

```python
# Remove all ',' 
trade['Export'] = trade['Export'].str.replace(',','')
trade['Import'] = trade['Import'].str.replace(',','')
trade['Total Trade'] = trade['Total Trade'].str.replace(',','')
trade['Trade Balance'] = trade['Trade Balance'].str.replace(',','')
```

```python
# changed data type for calculation
trade['Export'] = trade['Export'].astype('float')
trade['Import'] = trade['Import'].astype('float')
trade['Total Trade'] = trade['Total Trade'].astype('float')
trade['Trade Balance'] = trade['Trade Balance'].astype('float')
trade.dtypes
```

### Trends in Imports and Exports

```python
top_ten = trade.nlargest(10, ['Export'])
# nlargest method returns a specified number of rows ,starting from top after sorting the data frame
```

```python
sns.relplot(
    data=top_ten, 
    x="Export", y="Import", hue="Country",
    height=5, aspect=2.4, 
    kind="line"
).set(
    title="Total Trade", 
    ylabel="Import",
    xlabel="Export"
)
# hue is used to visualize the data of different categories in one plot
```

```python
top_ten = trade.nlargest(10, ['Import'])
```

```python
sns.relplot(
    data=top_ten, 
    x="Import", y="Export", hue="Country",
    height=5, aspect=2.4, 
    kind="line"
).set(
    title="Total Trade", 
    ylabel="Export",
    xlabel="Import"
)
```
